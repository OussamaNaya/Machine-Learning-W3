# Linear Regression

Linear regression uses the relationship between the data-points to draw a straight line through all them.
This line can be used to predict future values.

```python
import matplotlib.pyplot as plt

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.plot(x, y)
plt.show()
```

Import <b>scipy</b> and draw the line of Linear Regression:

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]


slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.title("Linear Regression", color="red", loc="center")
plt.show()
print("r = ",r)
```

# Regression Linear avec la method Trin/Test

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

train_x = x[:-5]
train_y = y[:-5]

test_x = x[-5:]
test_y = y[-5:]

def myfunc(x):
  return slope * x + intercept


slope, intercept, r, p, std_err = stats.linregress(train_x, train_y)
print("r_train = ",r)
mymodel = list(map(myfunc, train_x))

plt.scatter(train_x, train_y)
plt.plot(train_x, mymodel)
plt.title("Train")
plt.show()


slope, intercept, r, p, std_err = stats.linregress(test_x, test_y)
print("r_test = ",r)
mymodel = list(map(myfunc, test_x))

plt.scatter(test_x, test_y)
plt.plot(test_x, mymodel)
plt.title("Test")
plt.show()
```

```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)

plt.show()
```

<h3>R for Relationship</h3>

<li>It is important to know how the relationship between the values of the x-axis and the values of the y-axis is, if there are no relationship the linear regression can not be used to predict anything.</li>

<li>This relationship - the coefficient of correlation - is called r.</li>

<li>The r value ranges from -1 to 1, where 0 means no relationship, and 1 (and -1) means 100% related.</li>

<li>Python and the Scipy module will compute this value for you, all you have to do is feed it with the x and y values.,/li>

```python
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

print(r)
```

<b>Note:</b> The result -0.76 shows that there is a relationship, not perfect, but it indicates that we could use linear regression in future predictions.

<h3>Predict Future Values</h3>

<li>Now we can use the information we have gathered to predict future values.</li>

<li>Example: Let us try to predict the speed of a 10 years old car.</li>

<li>To do so, we need the same <b>myfunc()</b> function from the example above:

</li>


<h5>def myfunc(x):</h5>
  <h5>  return slope * x + intercept</h5>

```python
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

speed = myfunc(10)

print(speed)

#The example predicted a speed at 85.6, which we also could read from the diagram:
```
