# Étiquettes et titre Matplotlib

<h3>Créer des étiquettes pour un tracé : </h3>
    Avec Pyplot, vous pouvez utiliser les fonctions <p style="color:blue;">xlabel()</p> et ylabel()pour définir une étiquette pour les axes x et y.


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

 
<h3>Créer un titre pour une intrigue : </h3>
    <p>Avec Pyplot, vous pouvez utiliser la fonction <p style="color:red;">title()</p> pour définir un titre pour l'intrigue.</p>
    <p>&#128512;</p>

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

<h3>Définir les propriétés de police pour le titre et les étiquettes</h3>
Vous pouvez utiliser le fontdictparamètre dans <b>xlabel()</b>, <b>ylabel()</b> et <b>title()</b> pour définir les propriétés de police pour le titre et les étiquettes.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

<h3>Positionner le titre</h3>

Vous pouvez utiliser le <b>loc</b> paramètre in <b>title()</b> pour positionner le titre.

Les valeurs juridiques sont : <b>« left »</b>, <b>« right »</b> et <b>« centre »</b>. La valeur par défaut est « centre ».


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'right')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

# Sous-intrigue Matplotlib

<h3>Afficher plusieurs tracés</h3>
    Avec la <b>subplot()</b> fonction, vous pouvez dessiner plusieurs tracés sur une seule figure :
    <ul>
        <li>La mise en page est organisée en lignes et colonnes, représentées par le premier et le deuxième argument.</li>
    <li>Le troisième argument représente l'index du tracé actuel.</li>
    </ul>

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
```

```python
plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
```

```python
"""
    Ainsi, si nous voulons une figure avec 2 lignes et 1 colonne 
    (ce qui signifie que les deux tracés seront affichés l'un sur l'autre au lieu de côte à côte),
    nous pouvons écrire la syntaxe comme celle-ci :
"""
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

# Matplotlib Histograms
Un histogramme est un graphique montrant les distributions de fréquences .
Il s'agit d'un graphique montrant le nombre d'observations dans chaque intervalle donné.

<h3>Créer un histogramme</h3>
<li>Dans Matplotlib, nous utilisons la <b>hist()</b> fonction pour créer des histogrammes.</li>
<li>La <b>hist()</b> fonction utilisera un tableau de nombres pour créer un histogramme, le tableau est envoyé dans la fonction comme argument.</li>

# Graphiques circulaires Matplotlib


<h3>Création de diagrammes circulaires</h3>
Avec Pyplot, vous pouvez utiliser la fonction <b>pie()</b> pour dessiner des diagrammes circulaires :

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

<b>Remarque :</b> La taille de chaque coin est déterminée en comparant la valeur avec toutes les autres valeurs, en utilisant cette formule :

La valeur divisée par la somme de toutes les valeurs : <b>x/sum(x)</b>

<h3>Étiquettes</h3>
Ajoutez des étiquettes au diagramme circulaire avec le <b>labels</b> paramètre.

Le <b>labels</b> paramètre doit être un tableau avec une étiquette pour chaque coin :

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

<h3>Angle de départ</h3>
Comme mentionné, angle de départ par défaut est sur axe des x, mais vous pouvez modifier angle de départ en spécifiant un <b>startangle</b> paramètre.

Le <b>startangle</b> paramètre est défini avec un angle en degrés, angle par défaut est <b>0</b> :

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

<h3>Exploser</h3>
    <li>Peut-être voulez-vous que une des cales se démarque ? Le explodeparamètre vous permet de le faire.</li>
    <li>Le <b>explode</b> paramètre, il est spécifié, et non None, doit être un tableau avec une valeur pour chaque coin.</li>
    <li>Chaque valeur représente la distance par rapport au centre de chaque coin affiché :</li>

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.15, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

<h3>Ombre</h3>
Ajoutez une ombre au diagramme circulaire en définissant le <b>shadows</b> paramètre surTrue :

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

<h3>Couleurs</h3>
Vous pouvez définir la couleur de chaque coin avec le <b>colors</b> paramètre.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

<h3>Légende</h3>
Pour ajouter une liste explications pour chaque coin, utilisez la <b>legend()</b> fonction :

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

<h3>Légende avec en-tête</h3>
Pour ajouter un en-tête à la légende, ajoutez le <b>title</b> paramètre à la legend fonction.

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
